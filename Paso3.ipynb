```python
import numpy as np
import matplotlib.pyplot as plt


def legendre_p(n, x):
    if n == 0:
        return np.ones_like(x)
    elif n == 1:
        return x
    else:
        return ((2*n-1)*x*legendre_p(n-1, x) - (n-1)*legendre_p(n-2, x)) / n


def chebyshev_t(n, x):
    if n == 0:
        return np.ones_like(x)
    elif n == 1:
        return x
    else:
        return 2*x*chebyshev_t(n-1, x) - chebyshev_t(n-2, x)


# Cargar solo las columnas x2 y y2 desde un archivo CSV
# Asegúrate de que esta ruta sea correcta
ruta = "datos3.csv"
datos = np.genfromtxt(ruta, delimiter=',', skip_header=1)
x2 = datos[:, 0]  # Columna x
y2 = datos[:, 1]  # Columna y

# Generar puntos x en el intervalo [-1, 1]
x = np.linspace(-1, 1, 100)

y_legendre = sum(coef * legendre_p(int(idx), x) for idx, coef in enumerate(y2))
y_chebyshev = sum(coef * chebyshev_t(int(idx), x)
                  for idx, coef in enumerate(y2))

# Graficar los polinomios
plt.figure(figsize=(10, 5))
plt.plot(x, y_legendre, label='Polinomio de Legendre')
plt.plot(x, y_chebyshev, label='Polinomio de Chebyshev')
plt.legend()
plt.xlabel('x')
plt.ylabel('P(x)')
plt.title('Polinomios de Legendre y Chebyshev')
plt.grid(True)
plt.show()
plt.show()
```
